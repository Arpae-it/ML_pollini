<h3>Imports</h3>

```python
import os
import random
import calendar
import datetime
import warnings
import logging
import ast
import json
import numpy as np
import pandas as pd
from datetime import timedelta
from collections import Counter
from google.cloud import bigquery
import plotly.express as px
from tqdm.auto import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
import comet_ml
from comet_ml import Experiment, Optimizer
import scipy.stats
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import pprint

from tensorflow import keras
from tensorflow.keras.preprocessing import timeseries_dataset_from_array

logging.getLogger("comet_ml").setLevel("ERROR")

warnings.filterwarnings("ignore")
tqdm.pandas()
```

<h3>Config</h3>

```python
# Config

PROJECT_ID = 'arpae-prod-ml'

# BigQuery
BQ_DATASET = 'SAMPLE_DATA'
JOINED_BQ_DATASET = 'JOINED_DATA'

# Layout
COLOR_PALETTE = px.colors.qualitative.Prism

# Const
PERIOD_INIT = '2022-01-01'
PERIOD_END = '2022-07-01' 

# Paths
CLUSTER_FILENAME = "../data/clustering_intervals.csv" 
MODEL_FEATURES_FILENAME = "../data/model_features.json"

# models
MODELS_FOLDER = "../models/final/"

# Params
SEQUENCE_LEN = 7
BATCH_SIZE = 1024
```

<h3>Methods</h3>

```python
def _run_query(client, query):
    df = client.query(query).to_dataframe()
    return df

def _read_table(client, project_id, dataset, table):
    query = "SELECT * FROM `{}.{}.{}` ".format(project_id, dataset, table)
    df = _run_query(client, query)
    return df

def _read_table_delta(client, project_id, dataset, table, date_col, init, end):
    query = "SELECT * FROM `{}.{}.{}` WHERE {} > '{}' AND {} < '{}' ".format(project_id, dataset, 
                                                                             table, date_col, init, date_col, end)
    df = _run_query(client, query)
    if 'reftime' in df.columns:
        df.sort_values(by='reftime', inplace=True)
    elif date_col in df.columns:
        df.sort_values(by=date_col, inplace=True)
    else:
        return None
    return df

def _plot_performance_over(df, column):
    perf_df = df.groupby(column).mean()[['error']].reset_index()
    plt.figure(figsize=(15, 4))
    plt.title("MAE over {}".format(column))
    plt.bar(perf_df[column], perf_df['error'])
    plt.xlabel(column)
    plt.ylabel('Mean Squared Error')
    plt.xticks(perf_df[column].values)
    plt.grid(alpha=.4)
    plt.ylim(0, 1)
    plt.show()
    
def _plot_performance_over_h(df, column):
    perf_df = df.groupby(column).mean()[['error']].reset_index()
    plt.figure(figsize=(15, 6))
    plt.title("MAE over {}".format(column))
    plt.barh(perf_df[column], perf_df['error'])
    plt.ylabel(column)
    plt.xlabel('Mean Squared Error')
    plt.grid(alpha=.4)
    plt.xlim(0, 1)
    plt.show()
```

<h3>1. Config</h3>

<h4>1.1 Config BigQuery</h4>

```python
# Setup Client

bq_client = bigquery.Client(project=PROJECT_ID)
bq_client
```

<h3>2. Load Models & Data</h3>

<h4>2.1 Read Prediction Data</h4>

```python
# Read SLIDING_WINDOWS_DATASET

sliding_windows_dataset_df = _read_table_delta(bq_client, PROJECT_ID, JOINED_BQ_DATASET, 
                                               "SLIDING_WINDOWS_DATASET", "date",
                                               PERIOD_INIT, PERIOD_END)
sliding_windows_dataset_df['date'] = sliding_windows_dataset_df['date'].astype("str")
print(sliding_windows_dataset_df.shape)
sliding_windows_dataset_df.head(3)
```

<h4>2.2 Read Clusters Data</h4>

```python
# Read Season Clusters

clusters_df = pd.read_csv(CLUSTER_FILENAME)
print(clusters_df.shape)
clusters_df.head(3)
```

```python
# Get Clusters

clusters = sorted(clusters_df.cluster.unique())
clusters
```

<h4>2.3 Load Models</h4>

```python
# Load Clusters Models

models = {}
for cluster_id in clusters:
    model_name = [name for name in os.listdir(MODELS_FOLDER) 
                  if int(name.split("Cluster-")[-1].replace(".h5", ""))==cluster_id][0]    
    model = keras.models.load_model(os.path.join(MODELS_FOLDER, model_name))
    models[cluster_id] = model
```

<h4>2.4 Read Models Features</h4>

```python
# Read Json

model_features = json.load(open(MODEL_FEATURES_FILENAME))
```

<h3>3. Prepare Test Data</h3>

```python
# Set data

data_df = sliding_windows_dataset_df.copy()
```

<h4>3.1 Join Clusters with data</h4>

```python
# Merge data

data_df = pd.merge(data_df, clusters_df, on=['station_id', 'pol_var_id'])
print(data_df.shape)
data_df.head(3)
```

<h4>3.2 Preprocess Data</h4>

```python
features_plain = [
    'B13011_min_amin', 'B13011_max_amax', 'B13011_mean_mean', 'B13011_std_mean', 'B13011_sum_sum', 
    'B14198_min_amin', 'B14198_max_amax', 'B14198_mean_mean', 'B14198_std_mean', 'B14198_sum_sum',
    'TEMP_min_amin', 'TEMP_max_amax', 'TEMP_mean_mean', 'TEMP_std_mean', 'TEMP_sum_sum', 
    'station_lat_amax', 'station_lon_amax', 'station_H_piano_strada_amax', 'station_H_mslm_amax', 
    'week_amax', 
    'PREC_amin', 'PREC_mean', 'PREC_std', 'PREC_median', 'PREC_amax', 'PREC_skew', 'PREC_kurtosis',
]

features_paired = [
    'pol_value_amin', 'pol_value_mean', 'pol_value_std', 'pol_value_median',
    'pol_value_amax', 'pol_value_skew', 'pol_value_kurtosis',
    'pol_value_prev_1', 'pol_value_prev_2', 'pol_value_prev_3',
    'pol_value_prev_4', 'pol_value_prev_5', 'pol_value_prev_6',
    'pol_value_prev_7', 'pol_value_label',
    'seasonal_mean', 'season_label', 'seasonal_prev_1',
    'trend_label', 'trend_mean', 'trend_prev_1',
    'residual_label', 'residual_mean', 'residual_prev_1'
]
```

```python
# Update Float cols

for col in data_df.columns:
    try:
        if data_df[col] is not 'date' and data_df[col].dtype != np.float64:
            data_df[col] = data_df[col].astype(np.float64)
    except:
        print("Not available in {}".format(col))
```

```python
# Normalize station, species features

data_df.date = pd.to_datetime(data_df.date)
data_df.set_index("date", inplace=True)
df_grouped = data_df.groupby(['station_id', 'pol_var_id'])
```

```python
# Normalize other columns

scaler = MinMaxScaler()
data_df[features_paired] = df_grouped[features_paired].progress_apply(lambda x:
                                                                      pd.DataFrame(scaler.fit_transform(x), 
                                                                                   columns=x.columns)).values
data_df[features_plain] = scaler.fit_transform(data_df[features_plain])
```

<h3>4. Run Prediction</h3>

For each cluster, load its model and predict values for each <station, bcode>.<br>

```python
# Create dataset for each cluster and <station, bcode>

data_df_list = []
for cluster in tqdm(sorted(clusters)):
    # Get Cluster Model
    model = models[cluster]
    # Filter for current cluster
    curr_cluster_df = data_df[data_df['cluster']==cluster]
    # for each station, bcode in current cluster
    for station_id, pol_var_id in curr_cluster_df[['station_id', 'pol_var_id']].drop_duplicates().values:
        # Filter for current station_id, pol_var_id
        curr_df = curr_cluster_df[(curr_cluster_df['station_id']==station_id) & 
                                  (curr_cluster_df['pol_var_id']==pol_var_id)]
        # If current station, pol_var_id are not enought in current cluster, skip
        if curr_df.shape[0] < SEQUENCE_LEN*2:
            continue        
        # Keep only station & bcode of current cluster
        all_features = model_features[str(cluster)]
        # Create dataset
        dataset = timeseries_dataset_from_array(curr_df[all_features], 
                                                curr_df['pol_value_label'], 
                                                sequence_length=SEQUENCE_LEN, 
                                                batch_size=BATCH_SIZE)        
        # Run Prediction df
        preds = model.predict(dataset, verbose=False).squeeze()
        # Add to data
        for i in range(len(preds), curr_df.shape[0]):
            preds = np.append(preds, 0.0)
        curr_df['preds'] = preds
        data_df_list.append(curr_df)

data_df_preds = pd.concat(data_df_list)
print(data_df_preds.shape)
data_df_preds.head(3)
```

<h4>5.2 Performance Over Clusters, Stations, Bcodes</h4>

```python
# Get errors

errors_df_rows = []
for cluster in tqdm(sorted(clusters)):
    # Get Cluster Model
    model = models[cluster]
    # Filter for current cluster
    curr_cluster_df = data_df[data_df['cluster']==cluster]
    # for each station, bcode in current cluster
    for station_id, pol_var_id in curr_cluster_df[['station_id', 'pol_var_id']].drop_duplicates().values:
        # Filter for current station_id, pol_var_id
        curr_df = curr_cluster_df[(curr_cluster_df['station_id']==station_id) & 
                                  (curr_cluster_df['pol_var_id']==pol_var_id)]
        # If current station, pol_var_id are not enought in current cluster, skip
        if curr_df.shape[0] < SEQUENCE_LEN*2:
            continue        
        # Keep only station & bcode of current cluster
        all_features = model_features[str(cluster)]
        # Create dataset
        dataset = timeseries_dataset_from_array(curr_df[all_features], 
                                                curr_df['pol_value_label'], 
                                                sequence_length=SEQUENCE_LEN, 
                                                batch_size=BATCH_SIZE)        
        # Run Prediction df
        error = model.evaluate(dataset, verbose=False)
        # Add to dict
        row = [int(cluster), int(station_id), pol_var_id, error]
        errors_df_rows.append(row)

errors_df = pd.DataFrame(errors_df_rows, columns=['cluster', 'station_id', 'pol_var_id', 'error'])
print(errors_df.shape)
errors_df.head(3)
```

```python
# Performance over clusters, stations, bcodes
    
_plot_performance_over(errors_df, 'cluster')
_plot_performance_over(errors_df, 'station_id')
_plot_performance_over_h(errors_df, 'pol_var_id')
```
