```python
# Read Methods

def _run_query(client, query): 
    df = client.query(query).to_dataframe()
    return df

def _read_table(client, project_id, dataset, table):
    query = "SELECT * FROM `{}.{}.{}` ".format(project_id, dataset, table)
    df = _run_query(client, query)
    return df

def _read_table_delta(client, project_id, dataset, table, date_col, init, end):
    query = "SELECT * FROM `{}.{}.{}` WHERE {} > '{}' AND {} < '{}' ".format(project_id, dataset, table, date_col, init, date_col, end)
    df = _run_query(client, query)
    if 'reftime' in df.columns:
        df.sort_values(by='reftime', inplace=True)
    elif date_col in df.columns:
        df.sort_values(by=date_col, inplace=True)
    else:
        return None
    return df

def _extract_date(x):
    return str(x)[:10]

def _set_var_descr(x):
    if x in b_codes:
        return b_codes[x]
    else:
        return x

def _create_station_init_end_widget(df):
    station_ids = df.station_id.sort_values().unique()
    station_wdgt = widgets.Dropdown(options=station_ids, description='Station Id:', layout={"width":"50%"})
    init_wdgt = widgets.DatePicker(description='Init Date', value=datetime.date(2021,5,1))
    end_wdgt = widgets.DatePicker(description='End Date', value=datetime.date(2021,5,5))
    return station_wdgt, init_wdgt, end_wdgt

def _create_pol_var_id_widget(df):
    pol_var_ids = df.pol_var_id.sort_values().unique()
    pol_wdgt = widgets.Dropdown(options=pol_var_ids, description='Pol var id:', layout={"width":"50%"})
    return pol_wdgt

def _create_station_id_widget(df):
    station_ids = df.station_id.sort_values().unique()
    station_wdgt = widgets.Dropdown(options=station_ids, description='Station Id:', layout={"width":"50%"})
    return station_wdgt
```

```python
# Comp Methods

def _corr(x, y):
    #return scipy.stats.pearsonr(x, y)[0]
    #return scipy.stats.spearmanr(x, y).statistic
    return scipy.stats.spearmanr(x, y).correlation

def _windows_corr(df, feats, col_name):        
    # Remove where there is no pollen
    df = df[df['pol_value']!=0]
    # Get correlation for each feat
    corrs = []
    for feat in feats:
        corr = np.round(np.abs(_corr(df[feat], df[col_name])), 2)
        corrs.append(corr)
    # Get mean correlation
    mean_corr = np.round(np.mean([np.abs(c) for c in corrs if not math.isnan(c)]), 2)
    df_corr = pd.DataFrame([corrs + [mean_corr]], columns=feats + ['mean'])
    return df_corr

def _tsd_corr(data_df, feats, col):
    df_corr_list = []
    for station_id in tqdm(data_df.station_id.unique()):
        for pol_var_id in data_df.pol_var_id.unique():
            # filter
            df = data_df[(data_df['station_id']==station_id) & (data_df['pol_var_id']==pol_var_id)]
            # get corr
            df_corr = _windows_corr(df, feats, col)
            df_corr_list.append(df_corr)
    df_corr = pd.concat(df_corr_list)
    df_corr.reset_index(drop=True, inplace=True)
    df_corr = pd.DataFrame(df_corr.mean()).T
    return df_corr
```

```python
# Plot Methods

def hex_to_RGB(hex_str):
    """ #FFFFFF -> [255,255,255]"""
    return [int(hex_str[i:i+2], 16) for i in range(1,6,2)]

def get_color_gradient(c1, c2, n):
    assert n > 1
    c1_rgb = np.array(hex_to_RGB(c1))/255
    c2_rgb = np.array(hex_to_RGB(c2))/255
    mix_pcts = [x/(n-1) for x in range(n)]
    rgb_colors = [((1-mix)*c1_rgb + (mix*c2_rgb)) for mix in mix_pcts]
    return ["#" + "".join([format(int(round(val*255)), "02x") for val in item]) for item in rgb_colors]

def _plot_tsd_station_bcode(df, station_id, pol_var_id, mode='additive'):
    # decompose
    curr_df = df[(df['station_id']==station_id) & (df['pol_var_id']==pol_var_id)]
    curr_df.set_index(curr_df['pol_var_date'], inplace=True)
    curr_df.sort_index(inplace=True)
    result = seasonal_decompose(curr_df['pol_value'], model=mode, period=365)
    # plot
    plt.figure(figsize=(15, 20))
    plt.subplot(4, 1, 1)
    plt.title("{} - {} Observed".format(station_id, pol_var_id))
    plt.plot(pd.to_datetime(curr_df.pol_var_date), result.observed)
    plt.subplot(4, 1, 2)
    plt.title("{} - {} Seasonal".format(station_id, pol_var_id))
    plt.plot(pd.to_datetime(curr_df.pol_var_date), result.seasonal)
    plt.subplot(4, 1, 3)
    plt.title("{} - {} Trend".format(station_id, pol_var_id))
    plt.plot(pd.to_datetime(curr_df.pol_var_date), result.trend)
    plt.subplot(4, 1, 4)
    plt.title("{} - {} Residual".format(station_id, pol_var_id))
    plt.plot(pd.to_datetime(curr_df.pol_var_date), result.resid)
    plt.show()

def _plot_corr(df_corr, title):
    fig = go.Figure()
    fig = px.bar(title=title,
                 x=df_corr.to_dict().keys(), 
                 y=df_corr.loc[0].values,
                 color_discrete_sequence=COLOR_PALETTE,
                 template="simple_white")
    fig.update_xaxes(title_text='feats', showgrid=True)
    fig.update_yaxes(title_text='val', showgrid=True)
    fig.show()
    
def _plot_trend_over_years(df):
    for station_id, pol_var_id in df[['station_id', 'pol_var_id']].drop_duplicates().values:
        # filter curr df
        curr_trend_df = df[(df['station_id']==station_id) & (df['pol_var_id']==pol_var_id)]
        curr_trend_df.sort_values(by='week', inplace=True)    

        # mean corr
        fig = go.Figure()
        fig = px.area(title="{} - {}: Trend & StdDev change over years".format(station_id, pol_var_id),
                      data_frame=curr_trend_df,
                      x='week', 
                      y='trend',                    
                      line_shape='spline',
                      hover_name='trend',
                      
                      template="simple_white",
                      height=300)
        fig.update_traces(marker=dict(size=6), mode="markers+lines")
        fig.update_xaxes(title_text='week', showgrid=True, tickmode='linear')
        fig.update_yaxes(title_text='value', range=[-1, 1], showgrid=True)    
        fig.show('notebook')
        
def _plot_period_scatter_heatmap(data_df, period):
    week_df = pd.DataFrame(data_df.groupby(['year', period])['pol_value_normalized'].sum()).reset_index()
    week_df.head(3)

    # Plot week trend over years
    plt.figure(figsize=(22, 5))
    plt.subplot(1, 2, 1)
    plt.title("{} Value over years".format(period))
    for i, year in enumerate(sorted(data_df['year'].unique())):
        curr_df = week_df[week_df['year']==year]
        plt.scatter(curr_df[period], curr_df['pol_value_normalized'], color=colors[i], s=3)
        plt.plot(curr_df[period], curr_df['pol_value_normalized'], label=year, color=colors[i])  
    plt.xticks(week_df[period].unique())
    plt.xlabel(period)

    # Plot week-year hetamap
    week_df_pivot = week_df.pivot('year', period, 'pol_value_normalized')
    plt.subplot(1, 2, 2) 
    plt.title("{}-year Value Heatmap".format(period))
    sns.heatmap(week_df_pivot, annot=False)
    plt.xlabel(period)
    plt.ylabel("year")
    plt.show()
    
def _plot_agg_series(title, agg_series, df_pivot, xlabel):
    plt.figure(figsize=(20, 4))
    plt.subplot(1, 2, 1)
    plt.title(title)
    plt.bar(agg_series.keys(), agg_series.values)
    plt.xticks(agg_series.keys().tolist())
    plt.xlabel(xlabel)

    plt.subplot(1, 2, 2)
    plt.title("Year-Month Diffs")
    sns.heatmap(df_pivot.diff())
    plt.xlabel(xlabel)
    plt.ylabel("year")
    plt.show()
    
```

<h3>5. Period Variations Over Years</h3>

<h4>5.1 Extract year-peaks for each station, bcode</h4>

For each year, we can extract the week and the month where there is the <b>max week-mean-value</b> for the current year.<br>
Hence, the dataset could be reduced to a df with a row for each:
- station_id
- pol_var_id
- year
- adding values for peak_month & peak_week

```python
# Per ogni specie, anno, mese, week prendo il valore max

peak_df = data_df.groupby(['station_id', 'pol_var_id', 'year', 'month', 'week'])['pol_value_normalized'].sum().reset_index()
peak_df = peak_df[peak_df['pol_value_normalized']>0]
print(peak_df.shape)
peak_df.head(2)
```

```python
# Get the max for each mean value

rows = []
for station_id, pol_var_id, year in tqdm(peak_df[['station_id', 'pol_var_id', 'year']].drop_duplicates().values):
    # get curr df
    df = peak_df[(peak_df['station_id']==station_id) & 
                 (peak_df['pol_var_id']==pol_var_id) & 
                 (peak_df['year']==year)]
    row = df[df['pol_value_normalized']==df['pol_value_normalized'].max()].values[0]
    rows.append(row)
        
peak_df = pd.DataFrame(rows, columns=peak_df.columns)
peak_df.rename(columns={'month': 'peak_month', 
                        'week': 'peak_week', 
                        'pol_value_normalized': 'peak_value'}, inplace=True)
print(peak_df.shape)
peak_df.head(5)
```

Our aim is to identify if the week/month of peak vary during the years.

```python
# Get the mean peak values for every year

mean_df = peak_df.groupby('year')[['peak_week', 'peak_month']].mean().reset_index()
print(mean_df.shape)
mean_df.head(3)
```

```python
# Plot the mean peak week/month for every year

plt.figure(figsize=(20, 7))
plt.subplot(1, 2, 1)
plt.title("Peak-Month mean over years")
plt.scatter(mean_df['year'], mean_df['peak_month'], label='month')
plt.plot(mean_df['year'], mean_df['peak_month'])
plt.legend()
plt.ylim(0, 12)
plt.xticks(mean_df['year'].unique(), rotation=90)
plt.yticks(np.arange(13))
plt.ylabel('month')
plt.grid()

plt.subplot(1, 2, 2)
plt.title("Peak-week mean over years")
plt.scatter(mean_df['year'], mean_df['peak_week'], label='week')
plt.plot(mean_df['year'], mean_df['peak_week'])
plt.legend()
plt.ylim(0, 52)
plt.xticks(mean_df['year'].unique(), rotation=90)
plt.ylabel('week')
plt.yticks(np.arange(53))
plt.grid()
plt.show()
```

```python
# For each year, get the peak diff from the previous year

mean_df[['peak_week_diff', 'peak_month_diff']] = mean_df[['peak_week', 'peak_month']].diff()
mean_df.head(3)
```

```python
# The mean of each col identify how much is the difference between a year and the previous year

k_week = np.round(mean_df['peak_week_diff'].mean(), 2)
k_month = np.round(mean_df['peak_month_diff'].mean(), 2)
print("Every year, the peaks occur earlier by:")
print("- {} of a week".format(k_week))
print("- {} of a month".format(k_month))
```

<b>Hence, it seems that every year the peak occur ~7 hours earlier.</b>

Linear regression estimate the line coefficient over years.

```python
# Linear Regression

x, y = mean_df['year'].values, mean_df['peak_month'].values
m, b = np.polyfit(x, y, 1)

# Create an array of x values for the regression line
regression_x = np.linspace(min(x), max(x), 100)
# Calculate the corresponding y values for the regression line
regression_y = m * regression_x + b

print("Regression Slope: {}".format(m))
print("Regression Intercept: {}".format(b))
```

```python
# KDE and coeff Plot

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
kde_df = peak_df.groupby('year')[['peak_week', 'peak_month', 'peak_value']].mean().reset_index()
plt.title("Scatter plot over year, month. Slope={}, Intercept={}".format(np.round(m, 2), np.round(b, 2)))
sns.scatterplot(data=kde_df, x="year", y="peak_month")
plt.plot(regression_x, regression_y, color='red', label='Linear Regression')
plt.ylim(0, 12)
plt.yticks(np.arange(13))
plt.legend()
plt.grid()

kde_df = peak_df.groupby('year')[['peak_week', 'peak_month', 'peak_value']].mean().reset_index()
plt.subplot(1, 2, 2)
plt.title("Kde plot over year, month")
sns.kdeplot(data=kde_df, x="year", y="peak_month", legend=False, fill=True)
plt.plot(regression_x, regression_y, color='red', label='Linear Regression')
plt.grid()
plt.ylim(0, 12)
plt.yticks(np.arange(13))
plt.legend()

plt.show()
```

<h4>5.3 Find the stations/bcode where the difference are greater</h4>

We can perform the same analysis considering each station/species.<br>
So we can identify which one are the most interesting cases.

```python
# Get the mean peak values for every station, bcode, year

mean_df = peak_df.groupby(['station_id', 'pol_var_id', 'year'])[['peak_week', 'peak_month']].mean().reset_index()
print(mean_df.shape)
mean_df.head(3)
```

```python
# Get for each station/bcode how many week by mean are diff from the prev year

mean_diff_dict = {}
for station_id, pol_var_id in mean_df[['station_id', 'pol_var_id']].drop_duplicates().values:
    # Filter
    df = mean_df[(mean_df['station_id']==station_id) & (mean_df['pol_var_id']==pol_var_id)]
    df.sort_values(by='year', inplace=True)
    # if less than 8 year, drop
    if df['year'].nunique() < 19:
        continue    
    # Get the mean of the diff between years
    mean_diff = np.round(df['peak_week'].diff().mean(), 2)
    mean_diff_dict[str(station_id)+"-"+pol_var_id] = mean_diff
    
mean_diff_dict = (sorted(mean_diff_dict.items(), key=lambda item: abs(item[1])))
mean_diff_dict.reverse()
```

```python
# Plot distribution

mean_abs_diff = np.round(np.mean([abs(x[1]) for x in mean_diff_dict]), 2)

plt.figure(figsize=(15, 4))
plt.title("Week Diff distribution; mean asb diff: {}".format(mean_abs_diff))
plt.hist([x[1] for x in mean_diff_dict], bins=100)
plt.grid()
plt.show()
```

```python
print("Considered Cases: {}".format(len(mean_diff_dict)))
print("Total Cases: {}".format(mean_df[['station_id', 'pol_var_id']].drop_duplicates().shape[0]))
```

We're considering only cases that have data for every year from 2000 to 2022.<br>
These cases are 326 on 463 total cases.<br>

<b>By mean the variation over years is 0.15; it means that every year they increase or decrease by ~25h.</b>

```python
# Plot top cases

for el in mean_diff_dict[:10]:
    try:
        station_id, pol_var_id = el[0].split("-")
        station_id = int(station_id)
        print("Station: {} bcode: {} Mean_diff:{}".format(station_id, pol_var_id, el[1]))

        # Get current week & month peak 
        curr_peak_df = peak_df[(peak_df['station_id']==station_id) & (peak_df['pol_var_id']==pol_var_id)]
        kde_df = curr_peak_df.groupby('year')[['peak_week', 'peak_month', 'peak_value']].mean().reset_index()

        # Build regression line
        x, y = curr_peak_df['year'].values, curr_peak_df['peak_week'].values                        

        # V1
        #m, b = np.polyfit(x, y, 1)
        # V2
        reg = LinearRegression().fit(x.reshape(-1, 1), y)
        m, b = reg.coef_[0], reg.intercept_

        regression_x = np.linspace(min(x), max(x), 100)
        regression_y = m * regression_x + b    

        # Plot Scatter with Regression
        plt.figure(figsize=(15, 5))
        plt.subplot(1, 2, 1)
        plt.title("Scatter plot over year, week. Slope={}, Intercept={}".format(np.round(m, 2), np.round(b, 2)))
        sns.scatterplot(data=kde_df, x="year", y="peak_week")
        plt.plot(regression_x, regression_y, color='red', label='Linear Regression')
        plt.ylim(0, 53)
        plt.yticks(np.arange(0, 53, 2))
        plt.legend()
        plt.grid()

        # Plot KDE with Regression
        plt.subplot(1, 2, 2)
        plt.title("Kde plot over year, week")
        sns.kdeplot(data=kde_df, x="year", y="peak_week", legend=False, fill=True)
        plt.plot(regression_x, regression_y, color='red', label='Linear Regression')
        plt.grid()
        plt.ylim(0, 53)
        plt.yticks(np.arange(0, 53, 2))
        plt.legend()
        plt.show()
        
    except Exception as e:
        print(e)
        break
```
