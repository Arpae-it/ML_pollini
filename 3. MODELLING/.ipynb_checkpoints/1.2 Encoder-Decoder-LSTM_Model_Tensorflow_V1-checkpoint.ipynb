# Encoder-Decoder LSTM Model
Notebook to create a baseline Model with dataset create in notebook 0.<br>
The most common apporach to the problem seems to be a Encoder-Decoder LSTM Model as described <a href="https://medium.com/@jeewonkim1028/article-summary-generation-with-encoder-decoder-rnn-a13a5858aae2">here</a>.<br>
The notebook do the following steps:
- Data Preparation:
    - Token Encoding
    - Padding
- Token Embedding
- Model Creation:
    - Encoder (LSTM)
    - Decoder (LSTM)
- Training
- Testing
- Evaluation

```python
# Config

PREP_DATASET_PATH = "../../DATA/dataset/text-summarization/CNN_dataset/prep_test.csv"

SOSTOK = 'sostok'
EOSTOK = 'eostok'
```

```python
# Methods

def _wrap_text(x):
    #return '_START_ '+ x + ' _END_'
    return SOSTOK + x + EOSTOK

# Remove non-alphabetic characters (Data Cleaning)
def _text_strip(x):
    x = re.sub("(\\t)", " ", str(x)).lower()
    x = re.sub("(\\r)", " ", str(x)).lower()
    x = re.sub("(\\n)", " ", str(x)).lower()
    # Remove _ if it occurs more than one time consecutively
    x = re.sub("(__+)", " ", str(x)).lower()
    # Remove - if it occurs more than one time consecutively
    x = re.sub("(--+)", " ", str(x)).lower()
    # Remove ~ if it occurs more than one time consecutively
    x = re.sub("(~~+)", " ", str(x)).lower()
    # Remove + if it occurs more than one time consecutively
    row = re.sub("(\+\++)", " ", str(x)).lower()
    # Remove . if it occurs more than one time consecutively
    x = re.sub("(\.\.+)", " ", str(x)).lower()
    # Remove the characters - <>()|&"',;?~*!
    x = re.sub(r"[<>()|&\[\]\'\",;?~*!]", " ", str(x)).lower()
    # Remove mailto:
    x = re.sub("(mailto:)", " ", str(x)).lower()
    # Remove \x9* in text
    x = re.sub(r"(\\x9\d)", " ", str(x)).lower()
    # Replace INC nums to INC_NUM
    x = re.sub("([iI][nN][cC]\d+)", "INC_NUM", str(x)).lower()
    # Replace CM# and CHG# to CM_NUM
    x = re.sub("([cC][mM]\d+)|([cC][hH][gG]\d+)", "CM_NUM", str(x)).lower()
    # Remove punctuations at the end of a word
    x = re.sub("(\.\s+)", " ", str(x)).lower()
    x = re.sub("(\-\s+)", " ", str(x)).lower()
    x = re.sub("(\:\s+)", " ", str(x)).lower()
    # Replace any url to only the domain name
    try:
        url = re.search(r"((https*:\/*)([^\/\s]+))(.[^\s]+)", str(x))
        repl_url = url.group(3)
        row = re.sub(r"((https*:\/*)([^\/\s]+))(.[^\s]+)", repl_url, str(x))
    except:
        pass
    # Remove multiple spaces
    x = re.sub("(\s+)", " ", str(x)).lower()
    # Remove the single character hanging between any two spaces
    x = re.sub("(\s+.\s+)", " ", str(x)).lower()
    return x
```

<h4>4.2 Create Model Inputs</h4>

Prima di creare il modello, si definiscono gli input (sia le features che le labels).<br>
Si definisce innanzitutto un vocabolario di TOT parole.<br>
Si definisce poi il numero massimo di parole in X e in Y.<br>
Il Tokenizer converte ogni singola parola di ogni articolo in un intero che la codifica;<br>
Si agguinge poi del padding in fondo in modo che ogni input (ogni articolo descritto tramite un vettore di interi ciascuno dei quali codifica una parola) sia della stessa dimensione.<br>

Si creano ora questi vettori (di X e Y) che andranno in input al modello.

```python
def _vectorize_input(tokenizer, texts, max_len):
    # convert texts into encoded sequences
    vec = tokenizer.texts_to_sequences(texts) 
    # pad adding 0 to reach max length
    vec = pad_sequences(vec,  maxlen=max_len, padding='post')
    return vec
```

<h3>5. Define & Train Model</h3>

<h4>5.1 Config</h4>

```python
# Config

LATENT_DIM = 300
EMBEDDING_DIM = 200
```

<h4>5.2 Define Model</h4>

```python
# DEFINE ENCODER

# Inputs
encoder_inputs = Input(shape=(MAX_ARTICLE, ))

# Embedding layer (NB: questo layer pu essere sostituito da GraphSAGE ad esempio)
enc_emb = Embedding(VOCAB_LEN, EMBEDDING_DIM, trainable=True)(encoder_inputs)

# Encoder LSTM 1
encoder_lstm1 = LSTM(LATENT_DIM, 
                     return_sequences=True,
                     return_state=True, 
                     dropout=0.4,
                     recurrent_dropout=0.4)
(encoder_output1, state_h1, state_c1) = encoder_lstm1(enc_emb)

# Encoder LSTM 2
encoder_lstm2 = LSTM(LATENT_DIM, 
                     return_sequences=True,
                     return_state=True, 
                     dropout=0.4,
                     recurrent_dropout=0.4)
(encoder_output2, state_h2, state_c2) = encoder_lstm2(encoder_output1)

# Encoder LSTM 3
encoder_lstm3 = LSTM(LATENT_DIM, 
                     return_state=True,
                     return_sequences=True, 
                     dropout=0.4,
                     recurrent_dropout=0.4)
(encoder_outputs, state_h, state_c) = encoder_lstm3(encoder_output2)
```

```python
# DEFINE DECODER

# Set up the decoder, using encoder_states as the initial state
decoder_inputs = Input(shape=(None, ))

# Embedding layer
dec_emb_layer = Embedding(VOCAB_LEN, EMBEDDING_DIM, trainable=True)
dec_emb = dec_emb_layer(decoder_inputs)

# Decoder LSTM
decoder_lstm = LSTM(LATENT_DIM, 
                    return_sequences=True,
                    return_state=True, 
                    dropout=0.4,
                    recurrent_dropout=0.2)
(decoder_outputs, decoder_fwd_state, decoder_back_state) = decoder_lstm(dec_emb, initial_state=[state_h, state_c])

# Dense layer: TimeDistributed allows to apply a layer to every temporal slice of an input
#decoder_dense = TimeDistributed(Dense(VOCAB_LEN, activation='softmax'))
decoder_dense = Dense(VOCAB_LEN, activation="softmax")
decoder_outputs = decoder_dense(decoder_outputs)
```

```python
# DEFINE MODEL

model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.summary()
```

```python
plot_model(model, show_shapes=True)
```

<h4>5.3 Define Loss & Optimizer</h4>

```python
# Compile

model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy')
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=2)
```

<h4>5.4 Train</h4>

Il modello prende in input [encoder_inputs, decoder_inputs], dove i decoder_inputs corrispondono alle Y.<br>
Il decoder prende infatti in input ogni elemento della sequenza, usa h per creare un output e lo confronta con l'elemento y.

```python
history = model.fit(
    x=[train_X, train_Y[:, :-1]],
    y=train_Y.reshape(train_Y.shape[0], train_Y.shape[1], 1)[:, 1:],
    epochs=2, # 50
    callbacks=[es],
    batch_size=100,
    validation_data=([val_X, val_Y[:, :-1]], 
                     val_Y.reshape(val_Y.shape[0], val_Y.shape[1], 1)[:, 1:]))
```

```python
# Show

pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```

<h3>6. Test Model</h3>

Il modello in fase di training prende in input [encoder_inputs, decoder_inputs] dove decoder_inputs sono le Y dato che stiamo usando il Teacher Forcing Mechanism.<br>
Ora in fase di predizione i decoder_inputs corrisponderanno all'input stesso (=decoder_inputs).<br>

Si riutilizza quindi il Encoder e Decoder addestrati per generare un nuovo output:
- si prende l'Encoder e si predicono i vettori c e h in output
- si inizia poi ad iterare sugli input e generare ogni volta un nuovo output, finch non si trova una stop-condition:
    - si predice il vettore di output, h e c
    - si estrae il token predetto (tramite softmax) dal vettore di output
    - si aggiornano i vettori h e c
    - se il token predetto  EOSTOK o abbiamo raggiunto la lunghezza massima dell'output, si esce

```python
# Convert int-sequence to summary

def seq2summary(vocab_index, vocab_index_reverse, input_seq):
    newString = ''
    for i in input_seq:
        if i != 0 and i != vocab_index[SOSTOK] and i != vocab_index[EOSTOK]:
            newString = newString + vocab_index_reverse[i] + ' '
    return newString


# Convert int-sequence to text

def seq2text(vocab_index_reverse, input_seq):
    newString = ''
    for i in input_seq:
        if i != 0:
            newString = newString + vocab_index_reverse[i] + ' '
    return newString

# Method to run a prediction from an input sequence

def pred_from_sequence(vocab_index, vocab_index_reverse, encoder_model, decoder_model, input_seq):
    
    # Encode the input as state vectors.
    (e_out, e_h, e_c) = encoder_model.predict(input_seq)

    # Generate empty target sequence of length 1
    target_seq = np.zeros((1, 1))

    # Populate the first word of target sequence with the start word.
    target_seq[0, 0] = vocab_index['sostok']

    stop_condition = False
    decoded_sentence = ''

    while not stop_condition:
        (output_tokens, h, c) = decoder_model.predict([target_seq]+[e_out, e_h, e_c])

        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_token = vocab_index_reverse[sampled_token_index]

        if sampled_token != 'eostok':
            decoded_sentence += ' ' + sampled_token

        # Exit condition: either hit max length or find the stop word.
        if sampled_token == 'eostok' or len(decoded_sentence.split()) >= MAX_SUMMARY - 1:
            stop_condition = True

        # Update the target sequence (of length 1)
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        # Update internal states
        (e_h, e_c) = (h, c)

    return decoded_sentence
```

```python
# Inference Models

# Encode the input sequence to get the feature vector
encoder_model = Model(inputs=encoder_inputs, outputs=[encoder_outputs, state_h, state_c])

# Decoder setup

# Below tensors will hold the states of the previous time step
decoder_state_input_h = Input(shape=(LATENT_DIM, ))
decoder_state_input_c = Input(shape=(LATENT_DIM, ))
decoder_hidden_state_input = Input(shape=(MAX_ARTICLE, LATENT_DIM))

# Get the embeddings of the decoder sequence
dec_emb2 = dec_emb_layer(decoder_inputs)

# To predict the next word in the sequence, set the initial states to the states from the previous time step
(decoder_outputs2, state_h2, state_c2) = decoder_lstm(dec_emb2, initial_state=[decoder_state_input_h, decoder_state_input_c])

# A dense softmax layer to generate prob dist. over the target vocabulary
decoder_outputs2 = decoder_dense(decoder_outputs2)

# Final decoder model
decoder_model = Model([decoder_inputs]+[decoder_hidden_state_input, decoder_state_input_h, decoder_state_input_c],
                      [decoder_outputs2]+[state_h2, state_c2])
```

```python
def decode_sequence(input_seq):
    
    # Encode the input as state vectors.
    (e_out, e_h, e_c) = encoder_model.predict(input_seq)

    # Generate empty target sequence of length 1
    target_seq = np.zeros((1, 1))

    # Populate the first word of target sequence with the start word.
    target_seq[0, 0] = target_word_index['sostok']

    stop_condition = False
    decoded_sentence = ''

    while not stop_condition:
        (output_tokens, h, c) = decoder_model.predict([target_seq]+[e_out, e_h, e_c])

        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_token = reverse_target_word_index[sampled_token_index]

        if sampled_token != 'eostok':
            decoded_sentence += ' ' + sampled_token

        # Exit condition: either hit max length or find the stop word.
        if sampled_token == 'eostok' or len(decoded_sentence.split()) >= max_summary_len - 1:
            stop_condition = True

        # Update the target sequence (of length 1)
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        # Update internal states
        (e_h, e_c) = (h, c)

    return decoded_sentence
```

```python
# To convert sequence to summary
def seq2summary(input_seq):
    newString = ''
    for i in input_seq:
        if i != 0 and i != target_word_index['sostok'] and i \
            != target_word_index['eostok']:
            newString = newString + reverse_target_word_index[i] + ' '

    return newString


# To convert sequence to text
def seq2text(input_seq):
    newString = ''
    for i in input_seq:
        if i != 0:
            newString = newString + reverse_source_word_index[i] + ' '

    return newString
```

```python
for i in range(0, 19):
    print ('Body:', seq2text(x_tr[i]))
    print ('Original title:', seq2summary(y_tr[i]))
    print ('Predicted title:', decode_sequence(x_tr[i].reshape(1, max_text_len)))
    print ('\n')
```
