# LSTM on Cluster
Notebook to perform some model on a single Cluster.<br>
Data read are from table ALL_METEO_FEATS_WEEK_SLIDING_POL_DAT that contains a sliding windows of:
- feats about last 7-days meteo values
- pollen value for the next day

Cluster associations are read from a local file: we have different cluster annotations made by different techniques.<br>
Starting from past data, we can apply a <b>DNN/RNN Model</b> to forecast the next week or day.

```python
# Config

PROJECT_ID = 'arpae-prod-ml'

# BigQuery
BQ_DATASET = 'SAMPLE_DATA'
JOINED_BQ_DATASET = 'JOINED_DATA'

# Const
COMMON_PERIOD_INIT = '2011-01-01'
COMMON_PERIOD_END = '2021-12-31' 
#COMMON_PERIOD_YEAR = '2021'

# Feats
METEO_FEATS = ['min', 'max', 'mean', 'std']
METEO_VARS = ['B13011', 'B14198', 'PREC', 'TEMP']
DATE_COL = 'date'
FEATS = ['week', 'year', 'station_H_piano_strada', 'station_H_mslm', 'arkimet_lat', 'arkimet_lon',
         'B13011_min', 'B13011_max', 'B13011_mean', 'B13011_std', 'B13011_sum',
         'B14198_min', 'B14198_max', 'B14198_mean', 'B14198_std', 'B14198_sum',
         'TEMP_min', 'TEMP_max', 'TEMP_mean', 'TEMP_std', 'TEMP_sum', 
         'PREC_min', 'PREC_max', 'PREC_mean', 'PREC_std', 'PREC_sum', 
         'POL_min', 'POL_max', 'POL_mean', 'POL_std', 'POL_sum']
         #'prev_pol_value']
LABEL_COL = 'pol_value' 

K_INTERVAL = 5

# Layout
COLOR_PALETTE = px.colors.qualitative.Prism

OUTPUT_CLUSTER_FILENAME = "../data/clustering_intervals.csv"
```

<h3>Methods</h3>

```python
# Read Methods

def _run_query(client, query): 
    df = client.query(query).to_dataframe()
    return df

def _read_table(client, project_id, dataset, table):
    query = "SELECT * FROM `{}.{}.{}` ".format(project_id, dataset, table)
    df = _run_query(client, query)
    return df

def _read_table_delta(client, project_id, dataset, table, date_col, init, end):
    query = "SELECT * FROM `{}.{}.{}` WHERE {} > '{}' AND {} < '{}' ".format(project_id, dataset, table, date_col, init, date_col, end)
    df = _run_query(client, query)
    if 'reftime' in df.columns:
        df.sort_values(by='reftime', inplace=True)
    elif date_col in df.columns:
        df.sort_values(by=date_col, inplace=True)
    else:
        return None
    return df
```

```python
# Process Methods

def _shift_week(x, week):
    date = x - timedelta(days=7*week)
    return date.strftime("%Y-%m-%d")

def _set_range_value(x, intervals):    
    for i, interval in enumerate(intervals):
        if x <= interval:
            return np.float32(i)
    return None

def _create_scaling_widget():
    wdgt = widgets.Dropdown(options=['min-max-linear', 'interval'], 
                            description='Select Feats:', layout={"width":"50%"})
    return wdgt

def _create_cluster_widget(clusters):
    cluster_wdgt = widgets.Dropdown(options=clusters, description='Cluster id:', layout={"width":"50%"})
    return cluster_wdgt
```

<h3>1. Config</h3>

<h4>1.1 Config BigQuery</h4>

```python
# Setup Client

bq_client = bigquery.Client(project=PROJECT_ID)
bq_client
```

<h3>2. Read Data</h3>

<h4>2.1 Read Cluster file</h4>

```python
clusters_df = pd.read_csv(OUTPUT_CLUSTER_FILENAME)
print(clusters_df.shape)
clusters_df.head(3)
```

<h4>2.2 Read Tables</h4>

<b>ALL_METEO_FEATS_WEEK_SLIDING_POL_DAT</b> joins meteo features of last 7-days with the next-day pollen value.

```python
# Read ALL_METEO_FEATS_POL_DAT

all_meteo_week_feats_sliding_pol_dat_df = _read_table_delta(bq_client, PROJECT_ID, JOINED_BQ_DATASET, 
                                                            "ALL_METEO_FEATS_WEEK_SLIDING_POL_DAT", "init_date",
                                                            COMMON_PERIOD_INIT, COMMON_PERIOD_END)
for date_col in ['init_date', 'end_date', 'pol_var_date']:
    all_meteo_week_feats_sliding_pol_dat_df[date_col] = all_meteo_week_feats_sliding_pol_dat_df[date_col].astype("str")
print(all_meteo_week_feats_sliding_pol_dat_df.shape)
all_meteo_week_feats_sliding_pol_dat_df.head(3)
```

<h3>3. Prepare Data</h3>

<h4>3.1 Prepare Data</h4>

```python
# Set data

data_df = all_meteo_week_feats_sliding_pol_dat_df.copy()
```

```python
# Select a single cluster 

clusters = sorted(clusters_df.cluster.unique())
cluster_wdgt = _create_cluster_widget(clusters)
display(cluster_wdgt)
```

```python
# Filter data for station, bcode in the selected cluster

filt_clusters_df = clusters_df[clusters_df['cluster']==cluster_wdgt.value][['station_id', 'pol_var_id']]
data_df = pd.merge(data_df, filt_clusters_df, how='right', on=['station_id', 'pol_var_id'])
print(data_df.shape)
data_df.head(3)
```

<h4>3.2 Create Dataset</h4>

```python
# Create dataset

data_df.sort_values(['station_id', 'pol_var_id', DATE_COL], inplace=True)
dataset_df = data_df[['station_id', 'pol_var_id', DATE_COL]  + FEATS + [LABEL_COL]]
print(dataset_df.shape)
dataset_df.head(3)
```

```python
# Set index and drop nan

dataset_df.set_index(DATE_COL, inplace=True)
dataset_df.dropna(inplace=True)
print(dataset_df.shape)
dataset_df.head(3)
```

<h3>4. Preprocess Data</h3>

<h4>4.1 Normalize Features</h4>

```python
# Normalize all cols except for pollen ones; Save max cols: multiply predicted values to get not-normalized values

cols_max = []
for col in FEATS:
    scaler = MinMaxScaler()
    dataset_df[col] = pd.DataFrame(scaler.fit_transform(dataset_df[[col]])).values
    cols_max.append(int(scaler.data_max_))
print("Max cols:", cols_max)
print(dataset_df.shape)
dataset_df.head(3)
```

<h4>4.2 Normalize Pollen Values</h4>

Dataset contains the previous pollen value as feature and the next pollen value as label.<br>
We can normalize with a min-max-scaler or in a range interval.

```python
scaling_wdgt = _create_scaling_widget()
display(scaling_wdgt)
```

##### TODO: ############## FORSE ORA DEVO RAGGRUPPARE PER B-CODE PER TROVARE I MIN-MAX...

```python

if scaling_wdgt.value == 'interval':
    min_max = dataset_df[LABEL_COL].agg(['min', 'max'])
    intervals = np.linspace(min_max.min(), min_max.max(), K_INTERVAL)
    dataset_df[LABEL_COL] = dataset_df[LABEL_COL].apply(lambda x: _set_range_value(x, intervals))
        
elif scaling_wdgt.value == 'min-max-linear':
    scaler = MinMaxScaler()
    dataset_df[LABEL_COL] = pd.DataFrame(scaler.fit_transform(dataset_df[[LABEL_COL]])).values

else:
    print("Error on widget.")
    
print(dataset_df.shape)
dataset_df.head(3)
```

<h3>5. Create Train, Val, Test Datasets</h3>

We can use:
- data from 2011 to 2019 included as training set
- data of 2020 as validation set
- data of 2021 as test set

```python
# Print period

print("Data from: {} to: {}".format(dataset_df.index.min(), dataset_df.index.max()))
```

```python
# Split df into train and test sets

train_df = dataset_df[dataset_df.index < pd.to_datetime('2019-12-31')]
val_df = dataset_df[(dataset_df.index > pd.to_datetime('2019-12-31')) & 
                    (dataset_df.index < pd.to_datetime('2020-12-31'))]
test_df = dataset_df[(dataset_df.index > pd.to_datetime('2020-12-31')) & 
                     (dataset_df.index < pd.to_datetime('2021-12-31'))]

train_df.shape, val_df.shape, test_df.shape
```

```python
# Split into feats and labels

train_X, train_y = train_df[FEATS + [LABEL_COL]].values, train_df[LABEL_COL]
val_X, val_y = val_df[FEATS + [LABEL_COL]].values, val_df[LABEL_COL]
test_X, test_y = test_df[FEATS + [LABEL_COL]].values, test_df[LABEL_COL]

train_X.shape, val_X.shape, test_X.shape
```

```python
# Reshape X-inputs to be 3D for LSTM [samples, timesteps, features]

train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
val_X = val_X.reshape((val_X.shape[0], 1, val_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))

train_X.shape, val_X.shape, test_X.shape
```

<h3>6. DNN Model</h3>

```python
# Create Model V2

input_layer = tf.keras.layers.Input(shape=(1, train_X.shape[2]))
x = tf.keras.layers.Dense(units=128)(input_layer)
x = tf.keras.layers.Dropout(0.4)(x)
x = tf.keras.layers.Dense(units=128)(x)
x = tf.keras.layers.Dropout(0.4)(x)
output_layer = tf.keras.layers.Dense(units=1)(x)
model = tf.keras.models.Model(inputs=input_layer, outputs=output_layer) 
```

```python
# Compile Model

model.compile(loss='mse', optimizer='adam')
model.summary()
```

```python
# Train Model

history = model.fit(train_X, train_y, 
                    epochs=10,
                    validation_data=(val_X, val_y), 
                    verbose=1, 
                    shuffle=False)
```

```python
# Plot History

plt.figure(figsize=(15, 5))
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

<h4>6.2 Evaluate Model</h4>

```python
# Final Error

preds = model.predict(test_X).squeeze()    
mae = np.mean(np.abs(preds-test_y))
print("Final Mean Average Error: {}".format(mae))
```

<h4>6.3 Predict</h4>

```python
# Predict & Assign to test_df: add values for missing days in preds

for i in range(len(preds), test_df.shape[0]):
    preds = np.append(preds, 0.0)
test_df['preds'] = preds
```

```python
# Plot some station & bcode

N_SAMPLE = 10
for station_id, pol_var_id in test_df[['station_id', 'pol_var_id']].drop_duplicates().sample(N_SAMPLE).values:
    curr_test_df = test_df[(test_df['station_id']==station_id) & (test_df['pol_var_id']==pol_var_id)]
    plt.figure(figsize=(15, 5))
    plt.title("{} - {}".format(station_id, pol_var_id))
    plt.plot(curr_test_df.index, curr_test_df['preds'], label='pred')
    plt.plot(curr_test_df.index, curr_test_df[LABEL_COL], label='truth')
    plt.legend()
    plt.show()
```

<h4>6.4 Permutation Feature Importances (<a href="https://www.kaggle.com/code/cdeotte/lstm-feature-importance">link</a>)</h4>

```python
# Get baseline & features mae

results = []
oof_preds = model.predict(test_X, verbose=0).squeeze() 
baseline_mae = np.mean(np.abs(oof_preds-test_y))
results.append({'feature':'BASELINE','mae':baseline_mae})           
for k in tqdm(range(len(FEATS))):
    # shuffle feature k
    save_col = test_X[:,:,k].copy()
    np.random.shuffle(test_X[:,:,k])
    oof_preds = model.predict(test_X, verbose=0).squeeze() 
    mae = np.mean(np.abs(oof_preds-test_y))
    results.append({'feature':FEATS[k],'mae':mae})
    test_X[:,:,k] = save_col
    
mae_df = pd.DataFrame(results)
mae_df = mae_df.sort_values('mae')
```

```python
# Plot

plt.figure(figsize=(15,10))
plt.title('Feature Importance')
plt.barh(mae_df.feature, mae_df.mae)
plt.show()
```

<h3>7. No Windows LSTM Model</h3>

```python
# Create Model

input_layer = tf.keras.layers.Input(shape=(1, train_X.shape[2]))
x = tf.keras.layers.LSTM(units=128, dropout=0.4, return_sequences=True)(input_layer)
x = tf.keras.layers.LSTM(units=128, dropout=0.4, return_sequences=True)(x)
output_layer = tf.keras.layers.Dense(units=1)(x)
model = tf.keras.models.Model(inputs=input_layer, outputs=output_layer) 
```

```python
# Compile Model

model.compile(loss='mse', optimizer='adam')
model.summary()
```

```python
# Train Model

history = model.fit(train_X, train_y, 
                    epochs=10,
                    validation_data=(val_X, val_y), 
                    verbose=1, 
                    shuffle=False)
```

```python
# Plot History

plt.figure(figsize=(15, 5))
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

<h4>7.2 Evaluate Model</h4>

```python
# Final Error

preds = model.predict(test_X).squeeze()    
mae = np.mean(np.abs(preds-test_y))
print("Final Mean Average Error: {}".format(mae))
```

<h4>7.3 Predict</h4>

```python
# Predict & Assign to test_df: add values for missing days in preds

for i in range(len(preds), test_df.shape[0]):
    preds = np.append(preds, 0.0)
test_df['preds'] = preds
```
